```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from data_preprocessing_final import initiate_data
```

```python
data = pd.read_csv('Ames_HousePrice.csv', index_col=0)
```

```python
train_data_linear, train_target_linear, test_data_linear, test_target_linear, train_data_tree, train_target_tree, test_data_tree, test_target_tree = initiate_data(data)
```

```python
train_data_linear
```

```python
train_data_linear = train_data_linear.drop('PID', axis = 1)
test_data_linear = test_data_linear.drop('PID', axis = 1)
train_data_linear
```

```python
svm_model = svm.SVR(gamma = 0.01, epsilon=0.001)
svm_model.fit(train_data_linear, train_target_linear)
print(svm_model.score(train_data_linear, train_target_linear))
print(svm_model.score(test_data_linear, test_target_linear))
```

```python
from sklearn import model_selection
grid_para_svm = [
    {'kernel': ['poly'],
     'degree': [1],
    'gamma': [0.1, 0.001],
    'epsilon' : [.1, .001]},
    {'kernel': ['rbf'],
     'gamma': [0.1, 0.0001],
    'epsilon' : [.1, .001]}]
```

```python
grid_search_svm = model_selection.GridSearchCV(svm_model, grid_para_svm, cv=3, return_train_score=True)
grid_search_svm.fit(train_data_linear, train_target_linear)
```

```python
grid_search_svm.best_params_
```

```python
grid_search_svm.best_score_
```
