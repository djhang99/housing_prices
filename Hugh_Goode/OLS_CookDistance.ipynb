```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from data_preprocessing_final import cleaning, dummify_func
```

```python
from sklearn.model_selection import train_test_split
```

```python
import statsmodels.api as sm 
```

```python
data = pd.read_csv('Ames_HousePrice.csv', index_col=0)
```

```python
cleaned = cleaning(data)
```

```python
cleaned
```

```python
x,y = dummify_func(cleaned)
```

```python
X_PID = x.PID
X_PID
```

```python
X = x.drop('PID', axis = 1)
X
```

```python
y = y.SalePrice
y
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_add_const = sm.add_constant(X)
ols = sm.OLS(y, X_add_const)
ans = ols.fit()
print(ans.summary())
```

```python
influence = ans.get_influence()
cooks = influence.cooks_distance
cooks
```
