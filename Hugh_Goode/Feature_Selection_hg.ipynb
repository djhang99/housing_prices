```python
import numpy as np
import pandas as pd
%matplotlib inline 

housing = pd.read_csv('Ames_HousePrice.csv', index_col=0)
```

```python
price        = housing['SalePrice']
```

```python
housing.dtypes.unique()
```

```python
category = housing.select_dtypes('object')
category
```

```python
category.isnull().sum(axis = 0)
```

```python
category[~category.MiscFeature.isnull()]
```

```python
category[~category.isnull().any(axis = 1)]
```

```python
category2 = category.fillna('0')
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
oe.fit(category2)
category_enc = oe.transform(category2)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

fs = SelectKBest(score_func=chi2, k='all')
fs.fit(category_enc, price)
X_train_fs = fs.transform(category_enc)
```

```python
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
```

```python
for i in range(len(fs.scores_)):
    if fs.scores_[i] > 1000:
        print('Feature %d: %f' % (i, fs.scores_[i]))
```

```python
cols = np.array(category.columns).reshape(43,1)
cols
```

```python
r2_score = fs.scores_.reshape(43,1)
r2_score
```

```python
f_rank = pd.DataFrame(np.hstack([cols, r2_score]))
f_rank = f_rank.set_axis(['feat', 'r2'], axis=1, inplace=False)
f_rank
```
