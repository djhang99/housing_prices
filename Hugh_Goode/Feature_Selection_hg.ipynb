```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
coef = np.array(lasso.coef_).reshape(233,1)
coef
```

```python
col_name_dum = np.array(cat_dum.columns).reshape(233,1)
col_name_dum
```

```python
f_rank = pd.DataFrame(np.hstack([col_name_dum, coef]))
f_rank = f_rank.rename({0 : 'feat', 1 : 'coef'}, axis = 1)
f_rank
```

```python
f_rank['abs_coef'] = abs(f_rank.coef)
f_rank
```

```python
f_rank.sort_values('abs_coef', axis = 0, ascending = False)
```

```python
housing_num = housing.select_dtypes('int64', 'float64')
housing_num
```

```python
housing_num.columns
```

```python
## Numeric Colums to convert
# MSSubClass, OverallQual, OverallCond, YearBuilt, YearRemodAdd, MoSold, YrSold

# How to handle MiscVal???
```

```python
housing_num = housing_num.drop(['SalePrice', 'MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'MoSold', 'YrSold', 'MiscVal'], axis = 1)
housing_num
```

```python
housing_num.isnull().sum(axis = 0)
```

```python
housing_num2cat = housing[['MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'MoSold', 'YrSold', 'MiscVal']]
housing_num2cat
```

```python
category3 = pd.concat([category2, housing_num2cat], axis = 1)
category3
```

```python
cat_dum = pd.get_dummies(category3, drop_first = True)
cat_dum
```

```python
full_dum_data = pd.concat([housing_num, cat_dum], axis = 1)
full_dum_data
```

```python
lasso = Lasso()
lasso.fit(full_dum_data, price)
lasso.score(full_dum_data, price)
```

```python
col_name_dum = np.array(full_dum_data.columns).reshape(259,1)
coef = np.array(lasso.coef_).reshape(259,1)
col_name_dum
```

```python
f_rank = pd.DataFrame(np.hstack([col_name_dum, coef]))
f_rank = f_rank.rename({0 : 'feat', 1 : 'coef'}, axis = 1)
f_rank['abs_coef'] = abs(f_rank.coef)
f_rank.sort_values('abs_coef', axis = 0, ascending = False)
```

```python
f_rank[f_rank.coef == 0]
```
