```python
housing_num_PID = housing_num['PID']
housing_num = housing_num.drop(['PID', 'SalePrice', 'MSSubClass', 'OverallQual', 'OverallCond', 'MoSold', 'YrSold', 'MiscVal'], axis = 1)
housing_num
```

```python
housing_num2cat = housing[['MSSubClass', 'OverallQual', 'OverallCond', 'MoSold', 'YrSold', 'MiscVal']]
housing_num2cat
```

```python
category = pd.concat([category, housing_num2cat.astype(str)], axis = 1)
category
```

```python
category.dtypes
```

```python
cat_dum = pd.get_dummies(category, drop_first = True)
cat_dum
```

```python
cat_dum.columns
```

```python
scaler = MinMaxScaler()
scaler.fit(housing_num)
housing_num_scaled = scaler.transform(housing_num)
housing_num_scaled = pd.DataFrame(housing_num_scaled, columns = housing_num.columns)
housing_num_scaled.shape
```

```python
housing_num_scaled
```

```python
full_dum_data = pd.concat([housing_num_PID, housing_num_scaled, cat_dum], axis = 1)
full_dum_data
```

```python
full_dum_data.to_csv('dum_scaled_data.csv')
```

```python
price.to_csv('price_target.csv')
```

```python
lasso = Lasso()
lasso.fit(full_dum_data, price)
lasso.score(full_dum_data, price)
```

```python
col_name_dum = np.array(full_dum_data.columns).reshape(333,1)
coef = np.array(lasso.coef_).reshape(333,1)
col_name_dum
```

```python
f_rank = pd.DataFrame(np.hstack([col_name_dum, coef]))
f_rank = f_rank.rename({0 : 'feat', 1 : 'coef'}, axis = 1)
f_rank['abs_coef'] = abs(f_rank.coef)
f_rank.sort_values('abs_coef', axis = 0, ascending = False)
```
