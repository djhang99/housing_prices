```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from data_preprocessing_final import initiate_data
from sklearn.svm import SVR
```

```python
housing = pd.read_csv('Ames_HousePrice.csv', index_col = 0)
housing = housing[housing['SaleCondition'] == 'Normal']
```

```python
# Shoutout to Jack for carrying that cleaning file

X_train_lin, y_train_lin, X_test_lin, y_test_lin, X_train_tree, y_train_tree, X_test_tree, y_test_tree = initiate_data(housing)
```

```python
X_train_lin = X_train_lin.drop('PID', axis = 1)
X_train = X_train_lin
X_train
```

```python
y_train = y_train_lin
X_test = X_test_lin
X_test = X_test.drop('PID', axis = 1)
y_test = y_test_lin
y_train
```

```python
svr = SVR()
```

```python
paramDict = {'C': np.linspace(0.2, 0.25, 5),
             'epsilon': np.linspace(0.005, 0.1, 10),
             'kernel': ['linear']}
```

```python
grid = GridSearchCV(svr, paramDict, cv = 5, n_jobs = -1, verbose = 3, return_train_score = True)
fit = grid.fit(X_train, y_train)
```

```python
fit.best_params_
```

```python
beast = fit.best_estimator_
beast
```

```python
beast.score(X_train, y_train)
```
