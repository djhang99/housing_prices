```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from data_preprocessing_final import initiate_data, cleaning
from sklearn.svm import SVR
```

```python
housing = pd.read_csv('Ames_HousePrice.csv', index_col = 0)
```

```python
train_data_linear, train_target_linear, test_data_linear, test_target_linear, train_data_tree, train_target_tree, test_data_tree, test_target_tree = initiate_data(housing)
```

```python
X_train = train_data_linear.drop('PID', axis = 1)
y_train = train_target_linear
X_train
```

```python
X_test = test_data_linear.drop('PID', axis = 1)
y_test = test_target_linear
X_test
```

```python
svr = SVR()
paramDict = {'C': np.linspace(0.2, 0.25, 5),
             'epsilon': np.linspace(0.01, 0.05, 10),
             'kernel': ['linear']}
grid = GridSearchCV(svr, paramDict, cv = 5, n_jobs = -1, verbose = 3, return_train_score = True)
fit = grid.fit(X_train, y_train)
```

```python
fit.best_estimator_
```

```python
fit.score(X_train, y_train)
```

```python
fit.score(X_test, y_test)
```
