```python
import pickle
from sklearn.model_selection import train_test_split
from data_preprocessing_final import initiate_data
import pandas as pd
```

```python
housing = pd.read_csv('Ames_HousePrice.csv', index_col = 0)
```

```python
train_data_linear, train_target_linear, test_data_linear, test_target_linear, train_data_tree, train_target_tree, test_data_tree, test_target_tree = initiate_data(housing)
```

```python
X_train1 = train_data_linear.drop('PID', axis =1)
y_train = train_target_linear
X_train2 = train_data_tree.drop('PID', axis = 1)
X_train2.columns
```

```python
X_test1 = test_data_linear.drop('PID', axis = 1)
test_target_linear
X_test1
```

```python
y_train
```

```python
X_test2 = test_data_tree.drop('PID', axis = 1)
y_test = test_target_tree

# making SalePrice as a dataframe 
y_test = y_test.to_frame()
X_test2
```

```python
svm = pickle.load(open('svm_model.pkl', 'rb'))
ridge = pickle.load(open('ridge_model.pkl', 'rb'))
lasso = pickle.load(open('lasso_model.pkl', 'rb'))
gbm = pickle.load(open('gbm_model.pkl', 'rb'))
cat = pickle.load(open('cat_model.pkl', 'rb'))
elasticnet = pickle.load(open('elasticnet_model.pkl', 'rb'))
```

```python
fit = lasso.fit(X_train1, y_train)
print(fit.score(X_train1, y_train))
print(fit.score(X_test1, y_test))

print()

print()

fit3 = elasticnet.fit(X_train1, y_train)
print(fit3.score(X_train1, y_train))
print(fit3.score(X_test1, y_test))
```

```python
lassopred = fit.predict(X_test1)
elaspred = fit3.predict(X_test1)
gbmpred = gbm.predict(X_test2)
catpred = cat.predict(X_test2)
svmpred = svm.predict(X_test1)
```

```python
y_test
```

```python
import numpy as np

y_test['lassopred'] = lassopred
y_test['elaspred'] = elaspred
y_test['catpred'] = catpred

y_test['svmpred'] = svmpred
y_test = np.exp(y_test)

y_test['gbmpred'] = gbmpred
y_test
```

```python
# Tenk you David for helping out
ridge = pd.read_csv('jung.csv', index_col = 0)
ridge.rename(columns={'0':'ridgepred'}, inplace=True)
ridge = np.exp(ridge)
```

```python
ridge
```

```python
pred = pd.concat([y_test, ridge], axis=1)
pred = pred[['SalePrice','lassopred', 'ridgepred','elaspred', 'catpred', 'gbmpred', 'svmpred']]
pred
```

```python
pred['overallpred'] = pred.iloc[:, 1:7].mean(axis=1)
pred
```

```python
from sklearn.metrics import r2_score
ytrue = pred["SalePrice"].tolist()
ypred = pred["overallpred"].tolist()
r2 = r2_score(ytrue, ypred)
r2
```

```python
pred['Pct Diff'] = pred[['SalePrice', 'overallpred']].apply(lambda row: (row.iloc[0]-row.iloc[1])/row.iloc[0]*100, axis=1)
pred[['lassopred', 
        'ridgepred', 
        'elaspred', 
        'catpred', 
        'gbmpred',
        'svmpred',
        'overallpred']] = pred[['lassopred', 
                                  'ridgepred', 
                                  'elaspred', 
                                  'catpred', 
                                  'gbmpred',
                                  'svmpred',
                                  'overallpred']].round(2)
pred
```
